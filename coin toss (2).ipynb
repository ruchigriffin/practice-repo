```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
def TossCoin (n,p):
    
    results= []
    for i in range (n):
        if np.random.rand()< p:
            results.append (0)
        else:
            results.append (1)
    return results

```

```python
x= TossCoin(10,.5)
print (x)
heads= x.count (0)
tails= x.count (1)
```
